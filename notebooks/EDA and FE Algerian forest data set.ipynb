```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df=dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
df= df.rename(columns={
    '1': "Day",
    '6': "Month",
    '2012': "Year",
    '29': "Temperature",
    '57': "RH",
    '18': "Ws",
    '0': "Rain",
    '65.7': "FFMC",
    '3.4': "DMC",
    '7.6': "DC",
    '1.3': "ISI",
    '3.4.1': "BUI",
    '0.5': "FWI", 
    'not fire   ': "Classes",
})          
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df[["Month","Day","Year","Temperature","RH","Ws"]]=df[["Month","Day","Year","Temperature","RH","Ws"]].apply(pd.to_numeric, errors='coerce').astype('Int64')
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=="object"]
```

```python
objects
```

```python
for i in objects:
    if i != 'Classes':
        df[i]=df[i].apply(pd.to_numeric, errors='coerce').astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv("Algerian_forest_fires_cleaned_dataset.csv",index=False)
```
